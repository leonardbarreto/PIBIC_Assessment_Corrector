```python
'''
Considerando resultado = vitoria
'''
limiar=0.3
limiteMaxAposta=6
valorPagoEmpate=4.33
valorPagoVitoria=1.36
retornoReal=0
melhorRetorno=0
print("\t(Vit., Emp)")

for valorApostadoVitoria in range(1,limiteMaxAposta):
    for valorApostadoEmpate in range(1,limiteMaxAposta):
        totalInvestido = valorApostadoEmpate+valorApostadoVitoria
        for vitoria in range(0,2):
            if (vitoria==1):
                ganhoEmpate=0
                ganhoVitoria=valorPagoVitoria*valorApostadoVitoria
            else:
                ganhoEmpate=valorPagoEmpate*valorApostadoEmpate
                ganhoVitoria=0
            
            RetornoEmpate = ganhoEmpate-(valorApostadoEmpate)
            RetornoVitoria = ganhoVitoria-(valorApostadoVitoria)
            retornoReal=RetornoEmpate+RetornoVitoria
            maiorGanhoPercentual=retornoReal/(valorApostadoEmpate+valorApostadoVitoria)     
            if (retornoReal>0) and (maiorGanhoPercentual>limiar):   
                print('[',"VIT" if vitoria==1 else "EMP",']\t(',valorApostadoVitoria,',',valorApostadoEmpate,') =>',round(retornoReal,2),'\t',round(maiorGanhoPercentual,2))
```
